```python
def difference(A, B):
    d = {}
    result = ''
    for x in B:
        if x not in d:
            d[x] = True

    for x in A:
        if x not in d:
            result += str(x) + ' '
    print(result)

t = int(input())
for i in range(t):
    input()
    A = [ int(x) for x in input().split()]
    B = [ int(x) for x in input().split()]
    difference(A, B)
```
