```python
t = int(input())
for _ in range(t):
    n,k = [int(j) for j in input().split()]
    a = [a for a in input().split()]
    
    a_dct = {a[i]: 0 for i in range(0, len(a))}
    
    flag = 0
    
    for i in a:
        a_dct[i] += 1
    
    flag = 0
    for key, value in a_dct.items():
        if value == k:
            flag = 1
            print(key)
            break
    if flag == 0:
        print(-1)
```

```python
#More effifcient

for _ in range(int(input())):
    n,k = (int(i) for i in input().split())
    a = [a for a in input().split()]
    d = dict.fromkeys(a,0)
    
    for i in a:
        d[i]+=1
    
    print(d)
    flag = 0
    for i in a:
        if d[i]==k:
            flag = 1
            print(i)
            break
    
    if flag == 0:
        print(-1)
```
